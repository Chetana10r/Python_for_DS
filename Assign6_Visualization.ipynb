3. Display the top 5 highest average points among all countries.

4. Plot number of wines by country of origin

```python
import pandas as pd  
import matplotlib.pyplot as plt

country_counts = data['country'].value_counts()

plt.bar(country_counts.index, country_counts.values)
plt.title('Number of Wines by Country')
plt.xlabel('Country')
plt.ylabel('Number of Wines')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

This bar chart shows the Number of Wines by Country. Key insights:

The United States dominates with over 50,000 wines, far more than any other country.
France, Italy, and Spain follow, reflecting their strong wine traditions.
Countries like Portugal, Chile, and Argentina also contribute significantly.
Toward the right, countries like Slovakia, Peru, and Egypt have much fewer wines.

5. Plot a graph based on Highest point of wines vs COuntry of origin
a.WHat is the rank of India with respect to quality of wines?

```python
import pandas as pd
import matplotlib.pyplot as plt

highest_points = data.groupby('country')['points'].max().sort_values(ascending=False)

plt.figure(figsize=(10,6))
plt.scatter(highest_points.index, highest_points.values, color='green', s=100)
plt.title('Highest Points of Wines by Country (Scatter Plot)')
plt.xlabel('Country')
plt.ylabel('Highest Points')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

india_rank = highest_points.rank(ascending=False, method='min').get('India')

if india_rank:
    print(f"India's rank by highest wine quality (points) is: {int(india_rank)}")
else:
    print("India is not present in the dataset.")
```

This scatter plot shows the highest wine scores by country. The top-rated wines (with scores up to 100 points) come from Australia, Italy, France, Portugal, and the US, indicating their excellence in wine quality. As we move toward the right, countries like Slovakia, Peru, and Egypt have lower highest scores, suggesting less globally acclaimed wines from those regions.

6. Plot wine-producing provinces of the world(Category) to the number of labels of wines they produce(number). Plot top 10 of them.

```python
province_counts = data['province'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(province_counts.values, labels=province_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Top 10 Wine-Producing Provinces (Pie Chart)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

This pie chart shows the top 10 wine-producing provinces. California is the biggest contributor with 45.5% of wine production. Other major provinces include Washington (10.8%), Bordeaux (7.5%), and Tuscany (7.4%). The remaining provinces contribute smaller portions. This means California dominates wine production compared to others.

8. Plot area chart for points.

```python
import matplotlib.pyplot as plt

sorted_points = data['points'].sort_values().reset_index(drop=True)

plt.figure(figsize=(10,6))
plt.fill_between(sorted_points.index, sorted_points, color='red', alpha=0.6)
plt.title('Area Chart of Wine Points')
plt.xlabel('Index')
plt.ylabel('Wine Points')
plt.grid(True)
plt.tight_layout()
plt.show()
```

This area chart shows the distribution of wine points (quality scores). It highlights that most wines have scores between 80 and 100. The chart gradually rises, showing that higher scores are more common. Overall, it indicates most wines are of good quality.

9. Plot prices vs points and share your observations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.regplot(x='price', y='points', data=data, scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title('Price vs Points of Wines (with Trend Line)', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Points (Quality Score)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph shows that as the price of wine increases, the quality score (points) also increases slightly, as shown by the red trend line. However, most wines are reasonably priced and still have good scores. This means high price doesn’t always guarantee better quality. So, good wines can be found at lower prices too.

10. Plot Contribution of each Province.

```python
import matplotlib.pyplot as plt

province_counts = data['province'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(province_counts.head(10), labels=province_counts.head(10).index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Provinces by Wine Production Contribution (Pie Chart)', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

The visualizations show that California dominates global wine production, contributing nearly half of the total from the top 10 provinces. The US also leads in the number of wines produced, followed by France and Italy. However, in terms of quality, countries like Australia, Italy, France, Portugal, and the US have wines rated with the highest points (up to 100), indicating that both volume and quality are concentrated in a few key regions. Overall, wine excellence is globally distributed, but production is heavily region-specific.
