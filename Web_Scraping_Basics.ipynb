# Selenium another library for web scraping

```python
import requests
from bs4 import BeautifulSoup
r=requests.get("https://en.wikipedia.org/wiki/GeeksForGeeks")
#Displays the response from the website in the form of status code
print(r)
```

# Display the data in the HTML form using Beautiful SOup

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

# Display only div tag

```python
print(soup.div)
```

```python
x=soup.find("p")
print(x)

x=soup.find_all("p")
print(x)

print(soup.find_all(["script","link"]))

#p for paragraph
```

```python
x=soup.find_all("p")
for i in x:
    print(i.text)
```

```python
x=soup.find_all("p",class_="w-50")
for i in x:
    print(i)
```
