```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get("https://pict.edu/placement/")
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml") #lxml is a parser. or we can use html parser
print(soup)
```

```python
first_table=soup.find("table")
```

```python
# Extract headers
headers = []
table_headers = first_table.find_all('th')

if table_headers:  # If <th> elements are found, extract headers
    for th in table_headers:
        headers.append(th.text.strip())
else:  # If no <th> elements exist, use the first row as headers
    first_row = first_table.find('tr')  # Get the first row
    headers = [td.text.strip() for td in first_row.find_all('td')]  # Extract headers from <td>

# Extract rows
rows = []
for tr in first_table.find_all('tr')[1:]:  # Skip the first row if used as headers
    row_data = [td.text.strip() for td in tr.find_all('td')]
    if row_data:
        rows.append(row_data)

# Create DataFrame
df = pd.DataFrame(rows, columns=headers)
```
